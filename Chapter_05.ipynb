## 7.จงอธิบายความแตกต่างระหว่างลิสต์และทัพเพิล

```python
a = 10
b = 20
print ("a : {} , b : {}". format(a,b))

#ผลที่คาดหวังคือ
# a:20 , b:10
```

```python
#คำตอบ
a = 10
b = 20 

tmp = a 
a = b
b = tmp

print ("a : {} , b : {}".format(a,b))
```

```python
a = 10
b = 20 

# parallel assigment
a,b = b,a

print ("a : {} , b : {}".format(a,b))
```
